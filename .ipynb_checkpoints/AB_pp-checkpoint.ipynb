```python
import pandas as pd
import numpy as np
import scipy.stats as stats

import matplotlib.pyplot as plt
from matplotlib.axes import Axes
import seaborn as sns

%matplotlib inline

sns.set()

%config InlineBackend.figure_format = 'retina'
```

### О данных
[Данные](https://www.kaggle.com/yufengsui/datacamp-project-mobile-games-a-b-testing) взяты у команды мобильной игры Cookie Cats.
<br><br>
По мере прохождения игры игроки сталкиваются с препятствиями (воротами), тогда у них появляется выбор: подождать некоторое время или совершить покупку во внутриигровом магазине и продолжить сразу же. В этом проекте мы проанализируем результат A/B-теста, в котором первые ворота в Cookie Cats были перенесены с 30-го уровня на 40-й. В частности, мы проанализируем влияние на удержание игроков и игровые раунды.
<br><br>
У нас есть данные 90 189 игроков, которые установили игру в ходе эксперимента. При установке игрок случайно попадал в контрольную (ворота на 30-м уровне), либо тестовую (ворота на 40-м уровне) группу. Данные представлены так:
<br>
* `userid` - уникальный номер каждого игрока
<br>
* `version` - группа, в которой оказался игрок (gate_30 или gate_40)
<br>
* `sum_gamerounds` - количество раундов, сыгранных игроком в течение одной недели после установки
<br>
* `retention_1` - вернулся ли человек в игру через один день после установки
<br>
* `retention_7` - вернулся ли человек в игру через семь дней после установки

### Объяснительный анализ
Загружаем данные, делаем первый осмотр:

```python
df = pd.read_csv('https://raw.githubusercontent.com/ahilesseliha/git-example/master/cookie_cats.csv')
df.head()
```

```python
df.info()
```

<br>
Смотрим статистику по группам:

```python
df.loc[df['version'] == 'gate_30', 'version'] = 'control'
df.loc[df['version'] == 'gate_40', 'version'] = 'test'
```

```python
df.groupby('version')['sum_gamerounds'].describe()
```

<br>
Замечаем аномальное максимальное значение в первой группе. Похоже на выброс. Проверяем:

```python
fig, axes = plt.subplots(figsize=(15, 5))

df.groupby('version')['sum_gamerounds'].plot()

plt.show()
```

```python
df.sum_gamerounds.sort_values(ascending=False)
```

<br>
Выброс найден. Удалим его и снова проверим график:

```python
df = df[df['sum_gamerounds'] != df.sum_gamerounds.max()]
```

```python
fig, axes = plt.subplots(figsize=(15, 5))

df.groupby('version')['sum_gamerounds'].plot()

plt.show()
```

```python
df.groupby('version')['sum_gamerounds'].describe()
```

<br>
Теперь всё должно быть в порядке. Можем рассмотреть данные подробнее. Посмотрим на распределение сыгранных раундов внутри наших групп:

```python
control = df[df['version'] == 'control']['sum_gamerounds']
test = df[df['version'] == 'test']['sum_gamerounds']
```

```python
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(15, 5))
fig.subplots_adjust(wspace=0.3)

axes[0].hist([control, test], label=['control', 'test'])
axes[0].legend(loc='upper right')

sns.boxplot(x="version", y="sum_gamerounds", data=df)


plt.show()
```

<br>
В обеих группах распределенее схожее: ненормальное, сильно скошенное влево. Определить, есть ли значимые различия, без теста не выйдет. Этим мы займёмся позже, а пока ещё немного посмотрим на данные. 
<br>
<br>
Как много раундов проходят пользователи?

```python
user_count_for_rounds = df.groupby('sum_gamerounds').userid.count().reset_index().rename(columns={'userid' : 'users'})
user_count_for_rounds.head(20)
```

```python
user_count_for_rounds_control = pd.DataFrame(control).reset_index() \
                               .groupby('sum_gamerounds').index \
                               .count().reset_index() \
                               .rename(columns={'index' : 'users'})

user_count_for_rounds_test = pd.DataFrame(test).reset_index() \
                               .groupby('sum_gamerounds').index \
                               .count().reset_index() \
                               .rename(columns={'index' : 'users'})
```

```python
fig, axes = plt.subplots(ncols=3, nrows=1, figsize=(15, 5))
fig.subplots_adjust(wspace=0.3)

axes[0].plot(user_count_for_rounds)
axes[0].legend(['sum_gamerounds', 'users'])
axes[0].set_title('Total')

axes[1].plot(user_count_for_rounds_control)
axes[1].legend(['s', 'u'])
axes[1].set_title('Control')

axes[2].plot(user_count_for_rounds_test)
axes[2].legend(['s', 'u'])
axes[2].set_title('Test')


plt.show()
```

<br>
Всё довольно логично. Большинство пользователей остаются в игре на один и больше раундов. Но с увеличением числа раундов, пользователей становится всё меньше. Это справедливо для всех пользователей, независимо от группы.
<br>
<br>
Теперь посмотрим на ситуацию с удержанием. Создадим небольшую таблицу с разбиением по группам, количеством удержанных пользователей через один и семь дней, коэффициентом удержания:
<br>

```python
ret_tab1 = pd.crosstab('1_' + df.version, df.retention_1)
ret_tab7 = pd.crosstab('7_' + df.version, df.retention_7)

ret_tab = pd.concat([ret_tab1, ret_tab7])
```

```python
ret_tab
```

```python
ret_tab.columns = ['False', 'True']
ret_tab['Total'] = ret_tab['False'] + ret_tab['True']
ret_tab['Rate'] = ret_tab['True'] / ret_tab['Total']
```

```python
ret_tab
```

<br>
Сложно сказать, есть ли различия в рейтинге удержания через один день, в недельном рейтинге различие более заметное. Проверим то и другое через z-тест пропорций:
<br>

```python
from statsmodels.stats.proportion import proportions_ztest

count = np.array([ret_tab['True'][0], ret_tab['True'][1]])
nobs = np.array([ret_tab['Total'][0], ret_tab['Total'][1]])

stat, pval = proportions_ztest(count, nobs, alternative='two-sided')

print(f'p-value = {pval}')
```

```python
count = np.array([ret_tab['True'][2], ret_tab['True'][3]])
nobs = np.array([ret_tab['Total'][2], ret_tab['Total'][3]])

stat, pval = proportions_ztest(count, nobs, alternative='two-sided')

print(f'p-value = {pval}')
```

### A/B тест
И выборочная и контрольная группы, как мы помним, распределены ненормально, распределения сильно скошены влево, похожи на логнормальные. В таком случае лучше использовать не t-тест, а тест Манна-Уитни:
<br>


```python
stat, pval = stats.mannwhitneyu(control, test, alternative = 'two-sided')

print(f'p-value = {pval}')
```

<br>
p-value очень близко к 0.05, но всё ещё больше. Учитывая уровень значимости 0.05, мы не можем отклонить нулевую гипотезу о равенстве истинных средних. 
<br>
<br>
Теперь обратимся к мощности нашего тест и узнаем, какой минимальный эффект мы вообще могли отследить. Для этого, предположив, что наши распределения подобны логнормальным, позапускаем симуляцию. 
<br>
<br>
Тысячу раз извлекаем выборки размером с наши из двух логнормальных распределений и искусственно добавляем эффект. Для каждой из тысячи пар проводим тест Манна-Уитни и заносим p-value в список. В конце строим гистограмму списка и высчитываем, в каком проценте случаев p-value меньше уровня 0.05, это и есть наша мощность. Регулируя эффект, мы постепенно найдем то его минимальное значение, которое мог отследить наш тест, с учетом уровня значимости 0.05, мощности 0.8 и заданного размера выборок:
<br>


```python
p_array = []
effect = 1.005

for i in range(1000):
    a = pd.DataFrame(np.random.lognormal(2, 1, 45000))
    b = pd.DataFrame(np.random.lognormal(2 * effect, 1, 45000))
    
    p = stats.mannwhitneyu(a, b, alternative = 'two-sided')
    p_array.append(p[1])

p_array = pd.DataFrame(p_array)

p_array.plot(kind='hist', bins=20)

plt.show()

power = p_array[p_array < 0.05].count()[0] / 1000
print(f'Power of a test = {power}')
```

<br>
<br>
Видим, например, что для эффекта в полпроцента тест недостаточно мощен. Немного поперебирав значения, находим нужный размер эффекта:
<br>


```python
p_array = []
effect = 1.01

for i in range(1000):
    a = pd.DataFrame(np.random.lognormal(2, 1, 45000))
    b = pd.DataFrame(np.random.lognormal(2 * effect, 1, 45000))
    
    p = stats.mannwhitneyu(a, b, alternative = 'two-sided')
    p_array.append(p[1])

p_array = pd.DataFrame(p_array)

p_array.plot(kind='hist', bins=20)

plt.show()

power = p_array[p_array < 0.05].count()[0] / 1000
print(f'Power of a test = {power}')
```


## Итоги
Изучив данные, мы узнали, что:
<br>
* Независимо от группы, большинство пользователей остаются в игре на один и больше раундов. Дальше, с увеличением числа раундов, число пользователей сокращается.
<br><br>
* Пользователи чаще возвращаются в игру, когда первые ворота расположены на тридцатом, а не на сороковом уровне. Этот не совсем очевидный вывод можно попробовать объяснить пресыщением: потребив слишком много контента, пользователи теряли интерес к игре, а в случаях, когда потребление было ограничено меньшим уровнем, были вынуждены брать перерыв и интерес удавалось сохранить.
<br><br>
* Что касается количества сыгранных раундов, значимое отличие между группами, если оно есть, не превышает 1%. Изучив распределения сыгранных раундов в тестовой и контрольной группах, мы провели двусторонний тест Манна-Уитни. Тест с данным размером выборок, уровнем значимости 0.05 и мощностью 0.8 не выявил значимых различий. Анализ мощности показал, что эффект, если он есть, не превышает 1%.
